**NBA Data Web Scrapping using python**

Importing the required libraries

```python
#importing libraries
import pandas as pd
import requests
pd.set_option('display.max_columns', None) # so we can see all columns in the table
import time
import numpy as np
```

Checking the URL from which we will be importing the data

```python
test_Url ='https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season=2023-24&SeasonType=Regular%20Season&StatCategory=PTS'
```

```python
r = requests.get(url=test_Url).json() # saving the url request in a json file so that it is easy to access and manage
```

```python
table_Header = r['resultSet']['headers'] # checking the resultset and the headers in the dataset that we got from the test url to see if we are getting the right data
```

```python
pd.DataFrame(r['resultSet']['rowSet'], columns=table_Header) # saving the data into a dataframe
```

Creating  temperory data frames to store data from different year and different season types and trying to merge all the data into one data frame

```python
tmp_df1 = pd.DataFrame(r['resultSet']['rowSet'], columns=table_Header)
tmp_df2 = pd.DataFrame({'Year':['2012-13' for i in range(len(tmp_df1))],
                         'Season_type':['Regular%20Season' for i in range(len(tmp_df1))]}) # saving the data for year '2012-13' and season type 'regular season' in this data frame
tmp_df2
tmp_df3 = pd.concat([tmp_df2, tmp_df1], axis=1)
tmp_df3 # displaying the dataframe to verify the result
```

```python
del tmp_df1,tmp_df2,tmp_df3 # deleting the temp data frames that we created to now generate the actual forloop to get all the data we need
```

Code for looping through the years and season type and extract data in an excel file

```python
df_cols = ['Year', 'Season_type'] + table_Header # assiging the columns that we need to add in the for loop
```

```python
pd.DataFrame(columns=df_cols) # displaying how the dataframe result headers will look like
```

```python
header = {
    'accept':'*/*',
    'accept-encoding' : 'gzip, deflate, br, zstd',
    'accept-language': 'en-GB-oxendict,en-US;q=0.9,en-IN;q=0.8,en;q=0.7',
    'connection' : 'keep-alive',
    'dnt': '1',
    'host': 'stats.nba.com',
    'origin' : 'https://www.nba.com',
    'referer': 'https://www.nba.com/',
    'sec-ch-ua' : '"Not)A;Brand";v="99", "Google Chrome";v="127", "Chromium";v="127"',
    'sec-ch-ua-mobile' : '?1',
    'sec-ch-ua-platform' : '"Android"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-site',
    'user-agent': 'Mozilla/5.0 (Linux; Android 13; SM-G981B) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Mobile Safari/537.36'

} # we try to get the header from the url we are trying to scrape so that the url does not block us out and it is considered as a valid api request
```

Code for Looping through the time frame and the season types to be scraped

```python
final_df = pd.DataFrame(columns=df_cols) # creating an empty dataframe with the columns that we need
season_types = ['Regular%20Season', 'Playoffs'] # defining the different type of season we need to scrape
years = ['2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23', '2023-24'] # defining the list of years we need to scrape

start_loop_time = time.time()  # time when we first run this loop

for y in years: # first loop to loop through the years
  for s in season_types: # second loop to loop through the seasons
    api_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=PerGame&Scope=S&Season='+y+'&SeasonType='+s+'&StatCategory=PTS' # modify the url and change the date and season type to the loop variables
    r = requests.get(url=api_url, headers=header).json() # converting the request to a json format
    tmp_df1 = pd.DataFrame(r['resultSet']['rowSet'], columns=table_Header)
    tmp_df2 = pd.DataFrame({'Year':[y for i in range(len(tmp_df1))],
                            'Season_type':[s for i in range(len(tmp_df1))]})
    tmp_df3 = pd.concat([tmp_df2, tmp_df1], axis=1)
    df = pd.concat([final_df,tmp_df3], axis=0)
    print(f'Finished scraping data for the {y} {s}.') # printing the status of the loop
    lag = np.random.uniform(low=5,high=40) # defining the lag time between requests
    print(f'Waiting {round(lag,1)} seconds') # pring the lag time introduced between each year and session
    time.sleep(lag) # introducing the lag time

print(f'Process completed! Total run time: {round((time.time()-start_loop_time)/60,2)}') # printing the total run time of the loop
final_df.to_excel('nba_player_data.xlsx', index=False) # exporting the final dataframe to an excel file
```
